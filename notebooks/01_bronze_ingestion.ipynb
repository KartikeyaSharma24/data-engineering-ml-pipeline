```python

# if your workspace uses the 'workspace' catalog:
table_name = "workspace.default.s_p_500_stock_prices_2014_2017"  # adjust to what UI shows

df_bronze = spark.table(table_name)

# (Optional) standardize types (sometimes numbers come in as strings)
from pyspark.sql.functions import col, to_date
from pyspark.sql.types import DoubleType, LongType

df_bronze = (
    df_bronze
    .withColumn("date", to_date(col("date")))
    .withColumn("open", col("open").cast(DoubleType()))
    .withColumn("high", col("high").cast(DoubleType()))
    .withColumn("low", col("low").cast(DoubleType()))
    .withColumn("close", col("close").cast(DoubleType()))
    .withColumn("volume", col("volume").cast(LongType()))
)

# Write Bronze as a Delta table in your catalog (not to DBFS path)
bronze_tbl = "workspace.default.s_p_500_stock_prices_2014_2017"  # choose your target name
(df_bronze.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(bronze_tbl))

display(spark.table(bronze_tbl))
```
